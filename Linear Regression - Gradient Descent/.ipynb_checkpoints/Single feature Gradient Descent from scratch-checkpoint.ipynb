```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def cost(data, m, c):
    cost = 0
    N = len(data)
    for i in data:
        x = i[0]
        y = i[1]
        cost += (1/N) * ((y - m*x -c)**2)
    return cost

def gradient_descent(data, learning_rate, n_iterations):
    m, c = 0, 0
    N = len(data)
    cost = []
    for j in range(n_iterations):
        slope_m = 0
        slope_c = 0
        for i in data:
            x = i[0]
            y = i[1]
            slope_m += (-2/N) * (y - m*x - c) * x
            slope_c += (-2/N) * (y - m*x - c)
        m -= learning_rate * slope_m
        c -= learning_rate * slope_c
        cost += cost(data, m, c)
        
        
    return m, c
```

```python
data = np.loadtxt("data.csv", delimiter=",")
plt.scatter(data[:, 0], data[:, 1])
plt.style.use("dark_background")
plt.show()
n_iterations = 10
learning_rate = .01
m, c = gradient_descent(data, learning_rate, n_iterations)
plt.plot([0, 70], [c, m*70 +c])
plt.show()
```
