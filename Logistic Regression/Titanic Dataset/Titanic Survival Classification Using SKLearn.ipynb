```python
from sklearn import datasets
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./data/train.csv")
df
```

```python
df.drop("Name", axis=1, inplace=True) #Name is not a feature for our model
df.drop("Ticket", axis=1, inplace=True) #Ticket is not a feature for our model
df.drop("Cabin", axis=1, inplace=True)
df
```

```python
print(len(df["Embarked"]=="S"))
print(df[df["Embarked"]=="S"]["Survived"].sum())

print(len(df["Embarked"]=="C"))
print(df[df["Embarked"]=="C"]["Survived"].sum())

print(len(df["Embarked"]=="Q"))
print(df[df["Embarked"]=="Q"]["Survived"].sum())

#Since, number of people survived is radically different for people from each port. 
# "Embarked" is an important feature 
N = df.index.stop
df["PortS"]=0
df["PortC"]=0
df["PortQ"]=0
for i in range(N):
    if(df.iloc[i, 6]=="S"):
        df.iloc[i, 8] = 1
    elif(df.iloc[i, 6]=="C"):
        df.iloc[i, 9] = 1
    else:
        df.iloc[i, 10] = 1
df.drop("Embarked", axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt
plt.style.use("dark_background")
plt.scatter(df["Fare"], df["Survived"])
plt.show()
#Fare is useless
df.drop("Fare", axis=1, inplace=True)
df
```

```python
plt.style.use("dark_background")
plt.scatter(df["Age"], df["Survived"])
plt.show()
#Age is useless
df.drop("Age", axis=1, inplace=True)
df
```

```python
def convertToClasses(i):
    if(i=='male'):
        return 1
    return -1
df.Sex = df.Sex.apply(convertToClasses)
df
```

```python
X = df.iloc[:, [0,1,3,4,5]]
Y = df.iloc[:, 2]
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X, Y)
```

```python
model.score(X, Y)
```
