```python
import numpy as np
from sklearn import model_selection
import matplotlib.pyplot as plt
```

```python
# data
X = np.array([1, 2, 3, 4, 5, 6, 7, 8, 10, 11, 14, 15, 18.5, 20 ])
Y = np.array([1.5, 3, 4, 5, 6, 8, 11, 11.1, 10, 11, 16.2, 18, 18.9, 19.9])

X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y)
```

```python
plt.scatter(X_train, Y_train)
plt.style.use('dark_background')
plt.show()
```

```python
def getM(X, Y):
    return ( (X*Y).mean() - (X.mean() * Y.mean()) ) / ( (X**2).mean() - (X.mean())**2 )

def getC(X, Y, m):
    return Y.mean() - m*(X.mean())

def fit(X, Y):
    # Y = mX + c
    m = getM(X, Y)
    c = getC(X, Y, m)
    return m,c

m,c = fit(X_train, Y_train)
```

```python
# Cost = Sum( (Y - Y_pred)^2 ) / N
def cost(X, Y, m, c):
    return ((Y - (m*X + c))**2).mean()
cst = cost(X, Y, m, c)
```

```python
def predict(X, m, c):
    return m*X + c
```

```python
plt.scatter(X_train, Y_train)
plt.style.use('dark_background')
plt.plot([0, 20], [c, predict(20, m, c)])
plt.title("Plot of Y vs X on training data")
plt.show()
```

```python
def score(X, Y):
    Y_pred = predict(X, m, c)
    score =  1 - (( ((Y_pred-Y)**2).sum() )/ ( ((Y - Y.mean())**2).sum() )) 
    return score

print("Score on training data", score(X_train, Y_train))
print("Score on testing data", score(X_test, Y_test))
print("Slope is", m)
print("y-intercept is", c)
print("Cost is", cst)
```
