```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#Load training data
data = np.genfromtxt("train.csv", delimiter=",")
Y_train = data[:, 10]
X_train = data[:, 0:10]
```

```python
# Load evaluation data
X_test =  np.genfromtxt("test.csv", delimiter=",")
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
print(X_test.shape)
Y_pred = model.predict(X_test)
```

```python
# Analysing how model performs on training data
pred = model.predict(X_train)
plt.scatter(Y_train, pred)
plt.style.use("dark_background")
plt.plot([0, 350], [0, 350])
plt.xlabel("True output of training data")
plt.ylabel("Predicted output on training data")
plt.title("Predicted vs True value")
plt.show()
```

```python
score = model.score(X_train, Y_train)

#Actual Results will be based on testing data.
print("Score on training data is", score)
```
